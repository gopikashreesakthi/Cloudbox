Singleton Membership Function

Screenshot 2025-10-29 105944.png

```python
import numpy as np
import matplotlib.pyplot as plt

def singleton_mf(x, a):
    """Singleton Membership Function"""
    return np.where(np.isclose(x, a), 1, 0)

# New parameter
a = 8     # Different center value

# Membership calculation
x_values = np.array([7, 8, 9])
for val in x_values:
    mu = singleton_mf(np.array([val]), a)[0]
    print(f"μ({val}) = {mu}")

# Plot
x = np.linspace(0, 12, 100)
y = singleton_mf(x, a)

plt.figure(figsize=(7,4))
plt.plot(x, y, linewidth=2, color='blue', label='Singleton MF')
plt.scatter([a], [1], color='red', s=100, label=f'μ({a}) = 1')
plt.title("Singleton Membership Function (a = 8)")
plt.xlabel("x")
plt.ylabel("Membership Grade μ(x)")
plt.ylim([-0.1, 1.1])
plt.legend()
plt.grid(True)
plt.show()
```

Triangular Membership Function

Screenshot 2025-10-29 105944.png

```python
import numpy as np
import matplotlib.pyplot as plt

def triangular_mf(x, a, b, c):
    """Triangular Membership Function"""
    y = np.zeros_like(x)
    for i in range(len(x)):
        if a < x[i] < b:
            y[i] = (x[i] - a) / (b - a)
        elif b <= x[i] < c:
            y[i] = (c - x[i]) / (c - b)
        elif x[i] == b:
            y[i] = 1
        else:
            y[i] = 0
    return y

# New parameter values
a, b, c = 3, 7, 12

# Step 1: Compute μ(x) for a given input
x_input = 10
mu_x = max(min((x_input - a)/(b - a), (c - x_input)/(c - b)), 0)
print(f"μ({x_input}) = {mu_x:.2f}")

# Step 2: Plot graph
x = np.linspace(0, 15, 300)
y = triangular_mf(x, a, b, c)

plt.figure(figsize=(7,4))
plt.plot(x, y, color='purple', linewidth=2, label=f'a={a}, b={b}, c={c}')
plt.scatter([x_input], [mu_x], color='red', s=80, label=f'μ({x_input})={mu_x:.2f}')
plt.title("Triangular Membership Function (a=3, b=7, c=12)")
plt.xlabel("x")
plt.ylabel("Membership Grade μ(x)")
plt.ylim([-0.1, 1.1])
plt.legend()
plt.grid(True)
plt.show()
```

Trapezoidal Membership Function

Screenshot 2025-10-29 105944.png

```python
import numpy as np
import matplotlib.pyplot as plt

def trapezoidal_mf(x, a, b, c, d):
    """Trapezoidal Membership Function"""
    y = np.zeros_like(x)
    for i in range(len(x)):
        if a < x[i] < b:
            y[i] = (x[i] - a) / (b - a)
        elif b <= x[i] <= c:
            y[i] = 1
        elif c < x[i] < d:
            y[i] = (d - x[i]) / (d - c)
        else:
            y[i] = 0
    return y

# New parameter values
a, b, c, d = 2, 5, 9, 12

# Step 1: Compute μ(x) for a given input
x_input = 10
if x_input < a or x_input > d:
    mu_x = 0
elif a <= x_input <= b:
    mu_x = (x_input - a) / (b - a)
elif b <= x_input <= c:
    mu_x = 1
else:
    mu_x = (d - x_input) / (d - c)
print(f"μ({x_input}) = {mu_x:.2f}")

# Step 2: Plot graph
x = np.linspace(0, 14, 300)
y = trapezoidal_mf(x, a, b, c, d)

plt.figure(figsize=(7,4))
plt.plot(x, y, color='green', linewidth=2, label=f'a={a}, b={b}, c={c}, d={d}')
plt.scatter([x_input], [mu_x], color='red', s=80, label=f'μ({x_input})={mu_x:.2f}')
plt.title("Trapezoidal Membership Function (a=2, b=5, c=9, d=12)")
plt.xlabel("x")
plt.ylabel("Membership Grade μ(x)")
plt.ylim([-0.1, 1.1])
plt.legend()
plt.grid(True)
plt.show()
```

Gaussian Membership Function

Screenshot 2025-10-29 105944.png

```python
import numpy as np
import matplotlib.pyplot as plt

def gaussian_mf(x, m, sigma):
    """Gaussian Membership Function"""
    return np.exp(-0.5 * ((x - m) / sigma) ** 2)

# New parameters
m = 6.5     # Center of the curve
sigma = 2.0 # Width

# Step 1: Compute μ(x) for a given input
x_input = 9
mu_x = np.exp(-0.5 * ((x_input - m) / sigma) ** 2)
print(f"μ({x_input}) = {mu_x:.4f}")

# Step 2: Plot graph
x = np.linspace(0, 14, 400)
y = gaussian_mf(x, m, sigma)

plt.figure(figsize=(7,4))
plt.plot(x, y, color='orange', linewidth=2, label=f'm={m}, σ={sigma}')
plt.scatter([x_input], [mu_x], color='red', s=80, label=f'μ({x_input})={mu_x:.4f}')
plt.title("Gaussian Membership Function (m=6.5, σ=2.0)")
plt.xlabel("x")
plt.ylabel("Membership Grade μ(x)")
plt.ylim([-0.1, 1.1])
plt.legend()
plt.grid(True)
plt.show()
```

Generalized Bell-Shaped Membership Function

Screenshot 2025-10-29 105944.png

```python
import numpy as np
import matplotlib.pyplot as plt

def gbell_mf(x, a, b, c):
    """Generalized Bell-Shaped Membership Function"""
    return 1 / (1 + np.abs((x - c) / a) ** (2 * b))

# New parameter values
a = 1.8   # width
b = 2.5   # slope
c = 7.0   # center

# Step 1: Compute μ(x) for a given input
x_input = 9
mu_x = 1 / (1 + abs((x_input - c) / a) ** (2 * b))
print(f"μ({x_input}) = {mu_x:.4f}")

# Step 2: Plot graph
x = np.linspace(0, 14, 400)
y = gbell_mf(x, a, b, c)

plt.figure(figsize=(7,4))
plt.plot(x, y, color='teal', linewidth=2, label=f'a={a}, b={b}, c={c}')
plt.scatter([x_input], [mu_x], color='red', s=80, label=f'μ({x_input})={mu_x:.4f}')
plt.title("Generalized Bell-Shaped Membership Function (a=1.8, b=2.5, c=7.0)")
plt.xlabel("x")
plt.ylabel("Membership Grade μ(x)")
plt.ylim([-0.1, 1.1])
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid_mf(x, a, c):
    """Sigmoid Membership Function"""
    return 1 / (1 + np.exp(-a * (x - c)))

# New parameter values
a = 0.9    # slope
c = 5.5    # center point

# Step 1: Compute μ(x) for a given input
x_input = 8
mu_x = 1 / (1 + np.exp(-a * (x_input - c)))
print(f"μ({x_input}) = {mu_x:.4f}")

# Step 2: Plot graph
x = np.linspace(0, 12, 400)
y = sigmoid_mf(x, a, c)

plt.figure(figsize=(7,4))
plt.plot(x, y, color='darkred', linewidth=2, label=f'a={a}, c={c}')
plt.scatter([x_input], [mu_x], color='blue', s=80, label=f'μ({x_input})={mu_x:.4f}')
plt.title("Sigmoid Membership Function (a=0.9, c=5.5)")
plt.xlabel("x")
plt.ylabel("Membership Grade μ(x)")
plt.ylim([-0.1, 1.1])
plt.legend()
plt.grid(True)
plt.show()
```

Left Right Membership Function

Screenshot 2025-10-29 105944.png

```python
import numpy as np
import matplotlib.pyplot as plt

def lr_mf(x, a, b, L, R):
    """Left-Right (L–R) Membership Function"""
    y = np.zeros_like(x)
    for i in range(len(x)):
        if x[i] < a:
            y[i] = np.exp(-0.5 * ((a - x[i]) / L) ** 2)
        elif a <= x[i] <= b:
            y[i] = 1
        else:
            y[i] = np.exp(-0.5 * ((x[i] - b) / R) ** 2)
    return y

# New parameters
a, b = 4.5, 8.5
L, R = 1.2, 2.0

# Step 1: Compute μ(x) for a given input
x_input = 10
if x_input < a:
    mu_x = np.exp(-0.5 * ((a - x_input) / L) ** 2)
elif a <= x_input <= b:
    mu_x = 1
else:
    mu_x = np.exp(-0.5 * ((x_input - b) / R) ** 2)
print(f"μ({x_input}) = {mu_x:.4f}")

# Step 2: Plot graph
x = np.linspace(0, 14, 400)
y = lr_mf(x, a, b, L, R)

plt.figure(figsize=(7,4))
plt.plot(x, y, color='darkcyan', linewidth=2, label=f'a={a}, b={b}, L={L}, R={R}')
plt.scatter([x_input], [mu_x], color='red', s=80, label=f'μ({x_input})={mu_x:.4f}')
plt.title("L–R (Left–Right) Membership Function (a=4.5, b=8.5, L=1.2, R=2.0)")
plt.xlabel("x")
plt.ylabel("Membership Grade μ(x)")
plt.ylim([-0.1, 1.1])
plt.legend()
plt.grid(True)
plt.show()
```

Two-Parameter Pi-shaped (π) Membership Function

```python
import numpy as np
import matplotlib.pyplot as plt

def pi1_mf(x, a, b):
    """Π₁ (Two-Parameter Pi) Membership Function"""
    return 1 / (1 + ((x - a) / b) ** 2)

# New parameters
a = 6     # Center (peak)
b = 2     # Width

# Step 1: Compute μ(x) for a given input
x_input = 9
mu_x = 1 / (1 + ((x_input - a) / b) ** 2)
print(f"μ({x_input}) = {mu_x:.4f}")

# Step 2: Plot graph
x = np.linspace(0, 14, 400)
y = pi1_mf(x, a, b)

plt.figure(figsize=(7,4))
plt.plot(x, y, color='darkorange', linewidth=2, label=f'a={a}, b={b}')
plt.scatter([x_input], [mu_x], color='red', s=80, label=f'μ({x_input})={mu_x:.4f}')
plt.title("Π₁ (Two-Parameter Pi) Membership Function (a=6, b=2)")
plt.xlabel("x")
plt.ylabel("Membership Grade μ(x)")
plt.ylim([-0.1, 1.1])
plt.legend()
plt.grid(True)
plt.show()
```

Four-Parameter Pi-shaped (π) Membership Function

Screenshot 2025-10-29 105944.png

```python
import numpy as np
import matplotlib.pyplot as plt

def pi2_mf(x, lw, lp, rp, rw):
    """Π₂ (Pi-shaped) Membership Function"""
    y = np.zeros_like(x)
    for i in range(len(x)):
        if x[i] < lp:
            y[i] = lw / (lp + lw - x[i])
        elif lp <= x[i] <= rp:
            y[i] = 1
        else:
            y[i] = rw / (x[i] - rp + rw)
    return y

# 💡 Adjusted parameters for clearer shape
lw, lp, rp, rw = 3, 5, 9, 4

# Step 1: Compute μ(x) for a specific input
x_input = 11
if x_input < lp:
    mu_x = lw / (lp + lw - x_input)
elif lp <= x_input <= rp:
    mu_x = 1
else:
    mu_x = rw / (x_input - rp + rw)
print(f"μ({x_input}) = {mu_x:.4f}")

# Step 2: Plot
x = np.linspace(0, 16, 400)
y = pi2_mf(x, lw, lp, rp, rw)

plt.figure(figsize=(8,4.5))
plt.plot(x, y, color='purple', linewidth=2.5, label=f'lw={lw}, lp={lp}, rp={rp}, rw={rw}')
plt.scatter([x_input], [mu_x], color='red', s=80, label=f'μ({x_input})={mu_x:.2f}')

# Add annotation for clarity
plt.axvline(lp, color='gray', linestyle='--', linewidth=1)
plt.axvline(rp, color='gray', linestyle='--', linewidth=1)
plt.text(lp-0.4, -0.08, 'lp', fontsize=9, color='gray')
plt.text(rp-0.4, -0.08, 'rp', fontsize=9, color='gray')

plt.title("Π₂ (Pi-Shaped) Membership Function (lw=3, lp=5, rp=9, rw=4)")
plt.xlabel("x")
plt.ylabel("Membership Grade μ(x)")
plt.ylim([-0.1, 1.1])
plt.xlim([0, 15])
plt.legend()
plt.grid(True)
plt.show()
```

S–Shaped Membership Function

Screenshot 2025-10-29 105944.png



```python
import numpy as np
import matplotlib.pyplot as plt

def s_shaped_mf(x, a, b):
    """S-Shaped Membership Function (verified formula)"""
    y = np.zeros_like(x)
    for i in range(len(x)):
        if x[i] <= a:
            y[i] = 0
        elif a < x[i] <= (a + b) / 2:
            y[i] = 2 * ((x[i] - a) / (b - a)) ** 2
        elif (a + b) / 2 < x[i] <= b:
            y[i] = 1 - 2 * ((x[i] - b) / (b - a)) ** 2
        else:
            y[i] = 1
    return y

# Example parameters
a, b = 2, 8  # Different from PPT
x_input = 6  # Test input

# Step 1: Compute μ(x) manually
if x_input <= a:
    mu_x = 0
elif a < x_input <= (a + b)/2:
    mu_x = 2 * ((x_input - a) / (b - a)) ** 2
elif (a + b)/2 < x_input <= b:
    mu_x = 1 - 2 * ((x_input - b) / (b - a)) ** 2
else:
    mu_x = 1
print(f"μ({x_input}) = {mu_x:.4f}")

# Step 2: Plot the function
x = np.linspace(0, 10, 400)
y = s_shaped_mf(x, a, b)

plt.figure(figsize=(7,4))
plt.plot(x, y, color='teal', linewidth=2, label=f'a={a}, b={b}')
plt.scatter([x_input], [mu_x], color='red', s=80, label=f'μ({x_input})={mu_x:.4f}')
plt.title("S–Shaped Membership Function (a=2, b=8)")
plt.xlabel("x")
plt.ylabel("Membership Grade μ(x)")
plt.ylim([-0.1, 1.1])
plt.legend()
plt.grid(True)
plt.show()
```

Open-Right Membership Function

Screenshot 2025-10-29 105944.png

```python
import numpy as np
import matplotlib.pyplot as plt

def right_open_linear_mf(x, alpha, beta):
    """Linear Open-Right Membership Function"""
    y = np.zeros_like(x)
    for i in range(len(x)):
        if x[i] < alpha:
            y[i] = 0
        elif alpha <= x[i] <= beta:
            y[i] = (x[i] - alpha) / (beta - alpha)
        else:
            y[i] = 1
    return y

# Example parameters
alpha, beta = 3, 7
x_input = 5

# Compute μ(x)
if x_input < alpha:
    mu_x = 0
elif alpha <= x_input <= beta:
    mu_x = (x_input - alpha) / (beta - alpha)
else:
    mu_x = 1
print(f"μ({x_input}) = {mu_x:.4f}")

# Plot
x = np.linspace(0, 12, 400)
y = right_open_linear_mf(x, alpha, beta)

plt.figure(figsize=(7,4))
plt.plot(x, y, color='darkorange', linewidth=2, label=f'α={alpha}, β={beta}')
plt.scatter([x_input], [mu_x], color='red', s=80, label=f'μ({x_input})={mu_x:.2f}')
plt.title("Open-Right Membership Function (Linear) (α=3, β=7)")
plt.xlabel("x")
plt.ylabel("Membership Grade μ(x)")
plt.ylim([-0.1, 1.1])
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def left_open_linear_mf(x, alpha, beta):
    """Linear Open-Left Membership Function"""
    y = np.zeros_like(x)
    for i in range(len(x)):
        if x[i] < alpha:
            y[i] = 1
        elif alpha <= x[i] <= beta:
            y[i] = (beta - x[i]) / (beta - alpha)
        else:
            y[i] = 0
    return y

# Example parameters
alpha, beta = 4, 8
x_input = 6

# Compute μ(x)
if x_input < alpha:
    mu_x = 1
elif alpha <= x_input <= beta:
    mu_x = (beta - x_input) / (beta - alpha)
else:
    mu_x = 0
print(f"μ({x_input}) = {mu_x:.4f}")

# Plot
x = np.linspace(0, 12, 400)
y = left_open_linear_mf(x, alpha, beta)

plt.figure(figsize=(7,4))
plt.plot(x, y, color='royalblue', linewidth=2, label=f'α={alpha}, β={beta}')
plt.scatter([x_input], [mu_x], color='red', s=80, label=f'μ({x_input})={mu_x:.2f}')
plt.title("Open-Left Membership Function (Linear) (α=4, β=8)")
plt.xlabel("x")
plt.ylabel("Membership Grade μ(x)")
plt.ylim([-0.1, 1.1])
plt.legend()
plt.grid(True)
plt.show()
```
